```python
#Input parameters:


csvReference="https://www.ida.liu.se/~olesy12/rp/data/ReferenceData.csv"
csvData="https://www.ida.liu.se/~olesy12/rp/data/scdata.txt"
```

```python
import pandas as pd
import numpy as np
import scanpy as sc
import refproj

ref =pd.read_csv(csvReference, index_col=0)
FineCellType=ref["FineCellType"].to_numpy()
MainCellType=ref.index
ref=ref.drop("FineCellType",1)
```

```python
ssData0=pd.read_table(csvData, index_col=0)
ssData0=ssData0.transpose()

#Filter zero genes
res=sc.pp.filter_genes(ssData0, min_counts=1, inplace=False)
ssData=ssData0.iloc[:,res[0].to_list()].copy()
del ssData0
```

```python
TrueCellTypes=np.array(['CD4']*2500+['Monocyte']*2500+['CD8']*2500+['B-cell']*2500+['NK-cell']*2500)
```

```python
Labs, CellT, CellP=refproj.rp(ssData, ref, FineCellType, modelFolder=None, trueLabels=TrueCellTypes, resolution=0.3, savePlots=True)
```

```python
df=pd.DataFrame({"TrueCellTypes":TrueCellTypes, "Label":Labs})
G=df.groupby(["TrueCellTypes", "Label"]).size().unstack()
G[pd.isna(G)]=0
print(G)
```
